### HTML(Hyper Text Markup Language)  
- Living Standard 살아 있는 표준 : 지속적으로 업데이트되는 표준
- 다른 언어보다 느슨한 문법 : 표준을 준수하지 않고 코드를 작성 하면 유지 보수 비용 문제  
### HTML 요소
##### Element(요소) : Content(내용), Start Tag(시작 태그), End Tag(종료 태그)
![HTML 구성 요소](http://www.tcpschool.com/lectures/img_html_tag_structure.png)  
##### Start Tag, End Tag
- 요소의 시작과 끝을 의미하는 태그
##### Void 요소
- End Tag가 없는 태그
- Start Tag 내에 /를 넣는 Self-Closing 방식 or Start Tag만 작성  




```python
<!-- Self-Closing 방식 -->
<input/>
<hr/>
```

##### Content와 Element
- Content : Start Tag와 End Tag 사이의 내용을 의미
- Element : Start Tag, End Tag, Content를 모두 합친 하나

##### Attributes(속성)
- 태그의 동작을 제어하기 위한 사용자 지정 값
- 여는 태그/시작 태그 안에 사용되는 특수용어  
- 일반적인 요소의 형태:
<태그 속성="값">content to be modified by the tag</태그>
- 태그에 속성을 부여하는 여러가지 방법 :

```python
<!-- 값을 넣지 않고 속성을 지정하는 방법 -->
<input disabled />

<!-- 따옴표 없이 값을 넣어 속성을 지정하는 방법 -->
<input value=hello />

<!-- 따옴표로 값을 넣어 속성을 지정하는 방법.
작은 따옴표, 큰 따옴표 모두 가능 -->
<input type="checkbox" />
```

### HTML 요소 분류
##### 인라인 요소와 블록 요소
1. 인라인(inline) 요소: 
    * 태그가 할당된 텍스트나 이미지의 크기에 맞는 필요한 공간만을 차지함.
    * 높이나 너비를 지정할 수 없고 줄 내부 어디서든 시작

2. 블록(block) 요소:
    * 태그가 시작되면 이전 요소와 상관없이 개행해 새로운 줄에서 시작.
    * 너비는 좌우 양쪽으로 부모 요소의 100%를 차지
    * 왼쪽에서 오른쪽으로 확장

![예제 코드](./ch2_17.JPG)

```python
    <!-- 인라인 요소 <span>, <button>, <img> -->
    <!-- 블록 요소 <div> <article> <form> -->

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <style>
    div{
      border: 1px solid blue;
    }
    span,
    em{
      border: 1px dashed red;
    }
  </style>
</head>
<body>
  <div>block 요소</div>
  <div>
    <div>block 요소 안의 block 요소.</div>
    <span>block 요소 안의 inline 요소</span>
  </div>
  <span>inline 요소 안의 <em>inline 요소</em></span>
</body>
</html>
```

- 블록 요소는 인라인 요소와 블록 요소를 모두 포함. 인라인 요소는 인라인 요소만을 포함.

##### 콘텐츠 모델(Contents Model)
- 요소가 어떤 콘텐츠를 표현할 수 있는지, 어떤 하위 요소를 가지는지에 따라 Contents Model로 HTML요소 분류  
![콘텐츠 모델](https://blog.kakaocdn.net/dn/cqs6pF/btqBUz75S1p/Lay8r3Dcy0nYRcuZXV6ry0/img.png)

##### Content
1. Metadata Content
    * 나머지 콘텐츠의 표시나 동작을 설정하거나 문서와 다른 문서와의 관계를 설정  
2. Flow Content
    * 본문에 사용되는 대부분의 요소  
3. Selectioning Content
    * 아웃라인 정의
    * Heading(제목) 요소와 Footer(바닥글) 요소의 범위(Scope)을 정함  
4. Phrasing Content
    * 문서 내의 텍스트를 의미하고 단락을 형성하는 요소  
5. Heading Content
    * 섹션의 헤더를 정의하는 요소
6. Embedded Content
    * 외부의 자원을 가져오거나 삽입할 때 사용하는 요소
7. Interactive Content
    * 유저와 상호작용을 위해 특별히 설계된 요소

### HTML 문서 골격 만들기
##### DOCTYPE 문서의 타입을 지정

```python
<!DOCTYPE html>
```

##### html
- `<html>` 태그는 문서의 루트 지점을 명시하는 태그이다.
- lang 속성을 지정하여 문서의 언어가 장치에서 일관되게 판단되도록 해야 함

```python
<!DOCTYPE html>
<html lang="ko">
    <!-- lang 속성에 한국 언어 식별자 "ko" 할당 -->
</html>
```

##### head
- `<head>`태그는 문서의 제목과 문서의 인코딩 형식 등을 지정
- 아래는 자주 사용되는 중요한 요소들이다

```python
<!-- title 요소는 문서의 이름(제목)을 나타낸다. 문서당 한 개의 제목 요소가 있어야 하며 사용자 북마크나 검색 엔진에서 식별하는 기준으로 사용됨 -->
<!DOCTYPE html>
<html lang="en">
<head>
  <title>Document</title>
</head>

</html>
```

```python
<!-- meta 태그는 메타 데이터를 다룬다. 메타 데이터는 '데이터에 의한 데이터', '어떤 목적으로 만들어진 데이터'를 의미. 
주로 기계가 읽고 이해하는 정보. -->
<head>
  <meta charset="UTF-8" />
</head>
```

```python
<head>
  <meta http-equiv="X-UA-Compatible" content="IE-Edge" />
</head>
```

```python
<!-- body 태그는 문서의 내용이 들어가고 일반적으로 head 태그 다음에 작성.  -->
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta http-equiv="X-UA-Compatible" content="IE=edge">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Document</title>
</head>
<body>
  <!-- 문서의 내용 -->
</body>
</html>
```

#### 시맨틱(Semantic)
- HTML 요소는 시맨틱하게 작성해야한다.
- Semantic : "의미의", "의미론적인"
- 즉 의미에 맞는 태그를 사용해 문서를 작성하는 것
- ex> 특별한 의미가 없는 `<div>` 태그를 대신해 `<section>, <article>, <footer>`처럼 의미 있는 태그를 상황에 따라 사용하는 것

- 시맨틱하게 HTML을 작성하면 개발자 입장에서 가독성이 높아지고 유지 보수가 쉬워진다.
- 콘텐츠 구조를 더 쉽게 이해할 수 있음

##### How to
- 태그들이 만들어진 목적을 이해하고 그것에 맞게 사용

1. `<h1>~<h6>`
    * 각 세션의 제목을 나타냄
    * 숫자가 높을수록 글자가 작아짐
2. `<header>`
    * 제목이나 대표 이미지가 들어가는 요소
3. `<footer>`
    * Sectioning Content의 바닥글로 쓰임
4. `<main>`
    * 페이지의 콘텐츠 영역
    * 페이지당 한 번 사용 `<body>` 아래 직접 추가
    * 다른 요소 내에 중첩되지 않아야 함
5. `<article>`
    * 요소 자체가 하나의 의미 있는 콘텐츠 블록 영역임
    * 단일 게시물을 나타낼 때 사용
    * 블로그 항목, 게시물, 기사, 위젯에 사용
6. `<section>`
    * article 태그와 유사하지만 페이지의 단일 부분을 그룹화하는 데에 유용
    * 일반 컨테이너 요소가 아니기 때문에 단순 스타일링을 위한 요소로 사용할 때는 div 태그를 사용하는 것을 권장
7. `<aside>`
    * 간접적으로 관련된 추가 정보를 포함하는 요소
    * `<nav>` 요소나 광고, 인용처럼 분리된 컨텐츠를 나타낼 때 사용
8. `<nav>`
    * 다른 페이지 또는 내 문서의 특정 영역(탐색 링크가 있는 섹션)으로 이동시키는 링크

#### SEO(Search Engine Optimization)
- 사이트를 찾기 쉽도록 개선하는 여러 노력을 검색 엔진 최적화(SEO)라고 부름
##### How To
1. 시맨틱하게 HTML 작성
    * 시맨틱하게 코드를 작성함으로써 검색 엔진과 크롤러는 각 영역이 어떤 의미의 정보를 담는지 기계적으로 읽을 수 있음
    * 페이지의 개요와 구성, 본문 등을 파악하고, 키워드 등을 이해해 검색 시 사이트가 노출되는 것에 도움을 줌
2. `<title>`을 놓치지 말고 적절하게 작성
    * 사이트의 이름.
    * 자주 바꾸거나 길이가 너무 길면 검색 엔진에서 불이익을 받음
3. `<meta name="description">`을 이용해 페이지 설명 남기기 
    * `<meta name="description" content="페이지에 대한 간단한 설명">`
4. `<meta charset="UTF-8"/>`를 사용해 인코딩 방식을 지정하자
    * 인코딩 방식을 지정하여 여러 브라우저에서 통일된 인코딩 방식으로 노출시킴
5. open graph, twitter 태그를 사용하여 외부 사용자 유인
    * og(open-graph) 태그는 페이스북에서 만든 프로토콜
    * 컨텐츠의 요약된 내용이 sns에 게시될 때 최적화된 데이터를 표시할 수 있도록 설정하는 것
